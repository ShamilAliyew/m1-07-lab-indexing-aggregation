```python
import pandas as pd

roster = pd.DataFrame({
    "student_id": ["S051","S052","S053","S054","S055","S016","S057","S058","S059","S060",
                   "S001","S012","S013","S014","S015","S016","S017","S018","S019","S020"],
    "status": ["active","inactive","active","inactive","active","active","inactive","active","inactive","active",
               "active","inactive","active","inactive","active","active","inactive","active","inactive","active"]
})

merged = scores.merge(roster, on='student_id', how='left')
merged
```

```python
missing_status_count = merged["status"].isnull().sum()
print("Missing statuses:", missing_status_count)
```

```python
avg_by_module = scores.groupby(['cohort','module'])['score'].mean()
avg_by_module
```

```python
avg_by_cohort = scores.groupby('cohort')['score'].mean()
avg_by_cohort
```

#### the average values are equal to the average value of each cohort in `avg_by_module` dataframe

```python
student_module_report = scores.pivot_table(index='student_id', columns='module',values='score', aggfunc='mean')
student_module_report
```

```python
rows_report, cols_report = student_module_report.shape
rows_original = scores['student_id'].nunique()
print(rows_original == rows_report)
```

```python
modules_original = set(scores['module'].unique())
modules_report = set(student_module_report.columns)

print(modules_original == modules_report)
```

```python
top_students = scores.groupby(['student_id','cohort'])['score'].mean()
top_students = top_students.reset_index(name = "avg_score")
top_students
```

```python
top_students = top_students.sort_values(['cohort','avg_score'],
                                        ascending=[True, False])
top_students
```
